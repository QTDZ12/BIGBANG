```python
scores=[]
# plt.figure(figsize=(20,10),dpi=80)
for i in range(5,11):
    if(i==9):
        score=Table.read_table('20211109-1-utf-8.csv')
        score1=Table.read_table('20211109-utf-8.csv')
        scores.append(score)
        scores.append(score1)
    elif(i!=10):
        score = Table.read_table('2021110' + str(i) + '-utf-8.csv')
        scores.append(score)
    else:
        score = Table.read_table('20211110-utf-8.csv')
        scores.append(score)


id=range(5,12)
huang_scores=[]
huang_sort=[]
li_scores=[]
li_sort=[]
name1=input("请输入第一个学生姓名:")
name2=input("请输入第二个学生的姓名：")
for i in scores:
    work_select=i.select('姓名','总分','班名次').sort("总分",descending="True")
    huang=work_select.where('姓名',are.containing(name1))
    li=work_select.where('姓名',are.containing(name2))
    huang_scores.append(int(huang[1]))#将数组类型的数据强制转换为int
    li_scores.append(int(li[1]))
    huang_sort.append(int(huang[2]))
    li_sort.append(int(li[2]))
```

```python
students=Table().with_columns("考试序号",id,name1,huang_scores,name2,li_scores)
#设置图片的大小
```

```python
plt.rcParams['font.family'] = ['Fangsong']
plt_x=[i for i in range(5,12)]
plt.xticks(plt_x)
plt.title(name1+"和"+name2+"各次考试年级排名对比")
plt.xlabel("考试")
plt.ylabel("排名")
plt.grid(True) #同时显示
plt.plot(id,huang_scores,linestyle=':',label=name1,color="blue",linewidth=3.0)
plt.plot(id,li_scores,linestyle='-.',label=name2,color="yellow",linewidth=3.0)
plt.legend(loc="upper right")
plt.show()
```
